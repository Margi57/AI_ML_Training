```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("/content/drive/MyDrive/practice/KNNAlgorithmDataset.csv")
df
```

```python
df.isnull().sum()
```

```python
df['diagnosis'].unique()
```

```python
df.drop(columns=['Unnamed: 32'],inplace=True)
df
```

```python
# X=df[['radius_mean','texture_mean','perimeter_mean','area_mean','smoothness_mean','compactness_mean','concavity_mean']]
X=df.copy()
X.drop(columns=['diagnosis'],inplace=True)
X
```

```python
y=df['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
standardscaler=StandardScaler()
X_train=standardscaler.fit_transform(X_train)
X_test=standardscaler.transform(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model=KNeighborsClassifier()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test[0].reshape(1, -1))
y_pred
```

```python
from sklearn.metrics import accuracy_score,r2_score
```

```python
accuracyscore=accuracy_score(y_test,y_pred)
accuracyscore
```
